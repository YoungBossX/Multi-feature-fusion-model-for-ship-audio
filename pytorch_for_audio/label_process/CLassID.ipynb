```python
import pandas as pd
import os
```

```python
def process_xlsx_to_label_csv(xlsx_file_path, output_path):
    df = pd.read_excel(xlsx_file_path)
    if 'Filename' not in df.columns:
        raise ValueError("xlsx文件中未找到'Filename'列")
    if 'Type' not in df.columns:
        raise ValueError("xlsx文件中未找到'Type'列")

    filename_col = df['Filename']
    type_col = df['Type']

    type_mapping = {
        'Fishboat': 0,
        'Trawler': 0,
        'Mussel boat': 0,
        'Tugboat': 0,
        'Dredger': 0,

        'Motorboat': 1,
        'Pilot ship': 1,
        'Sailboat': 1,

        'Passengers': 2,

        'Ocean liner': 3,
        'RORO': 3,

        'Natural ambient noise': 4
    }

    class_id = type_col.map(type_mapping).astype(int)

    result_df = pd.DataFrame({
        'Filename': filename_col,
        'ClassID': class_id
    })

    result_df.to_csv(output_path, index=False)
    print(f"处理完成！已保存到{output_path}")

    return result_df
```

```python
xlsx_file_path = r"E:\数据集\ShipEar\shipsEar_AUDIOS\shipsEar.xlsx"
output_path = "label.csv"

process_xlsx_to_label_csv(xlsx_file_path, output_path)
```
