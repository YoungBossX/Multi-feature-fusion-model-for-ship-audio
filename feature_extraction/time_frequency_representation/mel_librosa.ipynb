```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 音频文件路径
audio_file = r'E:\数据集\ShipEar\data_preprocessing\3_Frame_Windows_2s_50%\6__10_07_13_marDeCangas_Entra_18.wav'

waveform, sample_rate = librosa.load(audio_file, sr=22050)

frame_length = 2048
hop_length = 512
```

```python
filter_banks = librosa.filters.mel(sr=sample_rate, n_fft=frame_length, n_mels=128)

plt.figure(figsize=(12, 8))
librosa.display.specshow(filter_banks, sr=sample_rate, x_axis='linear', y_axis='mel')
plt.title('Mel Filter Banks')
plt.tight_layout()
plt.show()
```

```python
mel_spectrogram = librosa.feature.melspectrogram(y=waveform, sr=sample_rate, n_fft=frame_length, hop_length=hop_length, n_mels=128)

log_mel_spectrogram = librosa.power_to_db(mel_spectrogram, ref=np.max)

plt.figure(figsize=(12, 8))
librosa.display.specshow(log_mel_spectrogram, sr=sample_rate, hop_length=hop_length, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title('Log Mel Spectrogram')
plt.tight_layout()
plt.show()
```
