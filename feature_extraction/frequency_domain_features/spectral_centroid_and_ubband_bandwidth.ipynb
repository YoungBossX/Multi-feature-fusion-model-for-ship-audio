```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_spectral_centroid(spectral_centroid, sample_rate, hop_length):
    frame = len(spectral_centroid)
    t = librosa.frames_to_time(np.arange(frame), sr=sample_rate, hop_length=hop_length)
    plt.figure(figsize=(12, 8))
    plt.plot(t, spectral_centroid)
    plt.title('Spectral Centroid')
    plt.xlabel('Time (s)')
    plt.ylabel('Frequency (Hz)')
    plt.show()
```

```python
def plot_spectral_bandwidth(spectral_bandwidth, sample_rate, hop_length):
    frame = len(spectral_bandwidth)
    t = librosa.frames_to_time(np.arange(frame), sr=sample_rate, hop_length=hop_length)
    plt.figure(figsize=(12, 8))
    plt.plot(t, spectral_bandwidth)
    plt.title('Spectral Bandwidth')
    plt.xlabel('Time (s)')
    plt.ylabel('Frequency (Hz)')
    plt.show()
```

```python
# 音频文件路径
audio_file = r'E:\数据集\ShipEar\data_preprocessing\3_Frame_Windows_2s_50%\6__10_07_13_marDeCangas_Entra_18.wav'

waveform, sample_rate = librosa.load(audio_file, sr=22050)

frame_length = 2048
hop_length = 512
```

```python
spectral_centroid = librosa.feature.spectral_centroid(y=waveform, sr=sample_rate, n_fft=frame_length, hop_length=hop_length,window='hamming')[0]
plot_spectral_centroid(spectral_centroid, sample_rate, hop_length)
```

```python
spectral_bandwidth = librosa.feature.spectral_bandwidth(y=waveform, sr=sample_rate, n_fft=frame_length, hop_length=hop_length,window='hamming')[0]
plot_spectral_bandwidth(spectral_bandwidth, sample_rate, hop_length)
```
