```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import hilbert
```

```python
def hilbert_envelope(waveform):
    """
    计算信号的希尔伯特变换并返回其 envelopes
    """
    # 计算信号的希尔伯特变换
    analytic_signal = hilbert(waveform)
    # 提取信号的 envelopes
    envelope = np.abs(analytic_signal)
    return envelope
```

```python
def plot_hilbert_envelope(waveform, waveform_HE, sample_rate, hop_length):
    """
    绘制信号的希尔伯特变换 envelopes
    """
    plt.figure(figsize=(12, 8))
    time_scale = librosa.samples_to_time(np.arange(0,len(waveform_HE)), sr=sample_rate)
    # 绘制原始波形（蓝色）
    librosa.display.waveshow(waveform, sr=sample_rate)
    # 绘制希尔伯特包络（红色）
    plt.plot(time_scale, waveform_HE, color='red')
    plt.title('Hilbert Transform Envelope of Audio Signal')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.show()
```

```python
# 音频文件路径
audio_file = r'E:\数据集\ShipEar\data_preprocessing\3_Frame_Windows_2s_50%\6__10_07_13_marDeCangas_Entra_18.wav'

waveform, sample_rate = librosa.load(audio_file, sr=22050)

frame_length = 2048
hop_length = 512

waveform_HE = hilbert_envelope(waveform)
plot_hilbert_envelope(waveform, waveform_HE, sample_rate, hop_length)
```
